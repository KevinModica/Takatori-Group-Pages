$\bf{\text{Spectral Methods are good for purely periodic or unbounded systems! }}$ It doesn't play nicely with no flux boundary conditions.

Example Code to Solve the Poisson equation in 1D, with periodic boundary conditions using Spectral Methods

\begin{equation}
\frac{\partial^2 f}{\partial x^2} = g(x)
\label{eq:1}
\end{equation}

Simple Idea. Fourier Transform Eq.1

\begin{equation}
-k^2 \hat{f}(k) = \hat{g}(k)
\end{equation}

Solve for $\hat{f}(k)$ and then invert the transform. Note that some places define k differently so check that if you need a factor of $2 \pi$

\begin{equation}
f(x) = - \mathscr{F}^{-1} \left[ \frac{\hat{g}(k)}{k^2} \right]
\end{equation}

$\color{red}{\text{Important Point!}}$ you get a singularity at k=0, so you have to artificially impose that $\hat{f}(k=0) = 0 $ which implies $\hat{g}(k=0) = 0$ and is peridoic I think

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

## Create function g = sin(x)
def g(x):
    out = np.sin(x)
    return out

t = np.linspace(-2*np.pi,2*np.pi,128) ## Sampling points locations
n = t.size ## Number of points
timestep = t[1] - t[0] ## distance between points

freq = 2*np.pi*np.fft.fftfreq(n, d=timestep) ## get the k-space equivalent to the points at t. Need factor of 2pi
b_k = np.fft.fft(g(t)) ## fast fourier transform of sin(x)
a_k = np.zeros(b_k.shape,dtype='complex')#  initialize f array, make sure you put type complex else it assumes float and screws things up.
a_k[0] = 0 + 0j# enforce that f(k=0) = 0 
a_k[1:] = - b_k[1:]/(freq[1:]*freq[1:])# Solve for f(k) 

f_x = np.fft.ifft(a_k) ## invert the transform

plt.plot(t,f_x.real) #plot numeric solution

plt.plot(t,-np.sin(t)) # plot analytic solution
```

Now let's try it in 2D

\begin{equation}
\frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} +  = \sin{x} + \cos{y}
\label{eq:1}
\end{equation}

Simple Idea. Fourier Transform Eq.1

\begin{equation}
-k_x^2 \hat{f}(k_x,k_y) -k_y^2 \hat{f}(k_x,k_y) = \hat{g}(k)
\end{equation}

Solve for $\hat{f}(k_x,k_y)$ and then invert the transform. Note that some places define k differently so check that if you need a factor of $2 \pi$

\begin{equation}
f(x) = - \mathscr{F}^{-1} \left[ \frac{\hat{g}(k)}{k_x^2 + k_y^2} \right]
\end{equation}

$\color{red}{\text{Important Point!}}$ you get a singularity at $k_x = k_y = 0$, so you have to artificially impose that $\hat{f}(k_x=0,k_y=0) = 0 $ which implies $\hat{g}(k_x=0,k_y=0) = 0$ and is peridoic I think

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Feb 17 17:28:18 2022

@author: kmodica
"""

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
pi = np.pi
## Create function g = sin(x)
def g(x,y):
    out = np.sin(x) +np.cos(y)
    return out

xs = np.linspace(-np.pi,np.pi,2**10) ## Sampling points locations
ys = xs
nx = xs.size ## Number of points
ny = ys.size
dx = xs[1] - xs[0] ## distance between points
dy = ys[1] - ys[0]

xx, yy = np.meshgrid(xs, ys,indexing='ij')
g_out = g(xx,yy)

xfreq = 2*np.pi*np.fft.fftfreq(nx, d=dx) ## get the k-space equivalent to the points at t. Need factor of 2pi
yfreq = 2*np.pi*np.fft.fftfreq(ny, d=dy) ## get the k-space equivalent to the points at t. Need factor of 2pi

b_k = np.fft.fft2(g_out) ## fast fourier transform of g(x,y)
a_k = np.zeros(b_k.shape,dtype='complex')#  initialize f array, make sure you put type complex else it assumes float and screws things up.

for ix in range(a_k.shape[0]):
    for iy in range(a_k.shape[1]):
        if ix ==0 and iy ==0:
            a_k[0,0] = 0 + 0j# enforce that f(k=0) = 0 
        else:
            a_k[ix,iy] = - b_k[ix,iy]/(xfreq[ix]**2 + yfreq[iy]**2)# Solve for f(k) 

f_xy = np.fft.ifft2(a_k) ## invert the transform

from matplotlib import cm

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})


# Plot the surface.
surf = ax.plot_surface(xx, yy, f_xy.real, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False)
plt.xlabel('x-axis')
ax.set_zlim(-2,2)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

analytic_F = -np.sin(xx) - np.cos(yy)
# Plot the surface.
surf = ax.plot_surface(xx, yy, analytic_F, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False)
ax.set_zlim(-2,2)
plt.xlabel('x-axis')

```
